### Introduction

Consider a Lotka Volterra system with inputs, i.e., a system defined by:
- the state variables $y_{prey}$ (preys) and $y_{pred}$ (predators),
- the inputs $u_{prey}$ (human intervention on the preys population) and $u_{pred}$ (human intervention on the predators population),
- the dynamics
$$
    \biggl \{

        \end{align}
$$

````python

#System:
#\dot{y}_{prey} &= \alpha y_{prey} - \beta y_{prey} y_{pred} - u_{prey} \\
#\dot{y}_{pred} &= - \gamma y_{pred} + \delta y_{prey} y_{pred} - u_{pred}

import sympy as sp

alpha = sp.Rational(1,5)  # 1/5
gamma = 1/30  # 1/3
u_prey = 1
u_pred = 0.2
beta = 0.1
delta = 0.3

y_prey, y_pred = sp.symbols('y_prey y_pred', real=True)

eq1 = sp.Eq(alpha*y_prey - beta*y_prey*y_pred - u_prey, 0)
eq2 = sp.Eq(-gamma*y_pred + delta*y_prey*y_pred - u_pred, 0)

solutions = sp.solve((eq1, eq2), (y_prey, y_pred), dict=True)

print(solutions)


# The only real solution is :

#0.154743986657038, y_prey: 5.41930221480947

````

### Question 5

You are worried that computing the equilibria is not enough to take a good decision, and you want to help the government by coding a simulator that can be used to draw predictions of how the population levels will change in time, depending on the interventions that humans may want to take in the future (and thus the $u$ signals in time). 

Code thus an Euler solver that computes the trajectories of the model, and plot the corresponding results, starting from generic initial conditions, and considering generic actuation signals $u$. Add as many cells you like to organize your code.

Tip: use as a starting point the code you find in https://github.com/damianovar/TTK4225-2023.

````python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

def myModel(y, t, u_preyT, u_predT):
    alpha = 1.1
    beta  = 0.7
    gamma = 0.4
    delta = 0.1

    yPrey, yPred = y

    dyPreydt = alpha * yPrey - beta * yPrey * yPred - u_preyT
    dyPreddt = -gamma * yPred + delta * yPrey * yPred - u_predT

    return np.array([dyPreydt, dyPreddt])

tMax = 70
dt = 0.02 
n = int(tMax / dt)

# Time array
t = np.linspace(0, tMax, n)

y = np.zeros((n, 2))
y[0] = [6.5, 1.73]  



# Assuming human intervention don't have a constant effect over time???
# If so, these functions can be replaced by constants
def uPreyFunc(t):
    return 0.1 * np.sin(0.1 * t)

def uPredFunc(t):
    return 0.05 * np.cos(0.1 * t)

for i in range(n - 1):
    u_preyT = uPreyFunc(t[i])
    u_predT = uPredFunc(t[i])
    dydt = myModel(y[i], t[i], u_preyT, u_predT)
    #Using euler integration
    y[i + 1] = y[i] + dt * dydt

# Plot results
plt.figure(figsize=(10, 6))

# Plot populations vs time
plt.subplot(2, 1, 1)
plt.plot(t, y[:, 0], label='Prey')
plt.plot(t, y[:, 1], label='Predator')
plt.title('Population Dynamics Over Time (Euler Method)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.grid()

# Plot phase portrait
plt.subplot(2, 1, 2)
plt.plot(y[:, 0], y[:, 1])
plt.title('Phase Portrait')
plt.xlabel('Prey Population')
plt.ylabel('Predator Population')
plt.grid()

plt.tight_layout()
plt.show()
````
